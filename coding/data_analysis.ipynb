```python
import pandas as pd
import json
import os
from collections import defaultdict
import numpy as np

from datetime import datetime, timedelta
import matplotlib.pyplot as plt
```

```python
demo_path = "../../html.2023.final.data/demographic.json"

with open(demo_path, 'r') as f:
    data = json.load(f)
```

### Total Area

```python
dic = defaultdict(int)
for station in list(data.values()):
    dic[station["sarea"]] += 1
dic["Total"] = sum(dic.values())
pd.DataFrame(dic, index=["站點數目"]).T
```

```python
folder_path = '../..//html.2023.final.data/release/'
dic_file_cnt = defaultdict(int)
for item in os.listdir(folder_path):
    item_path = os.path.join(folder_path, item)
    # print(item_path[-4:])
    # Get the list of files in the folder
    files = os.listdir(item_path)
    # Count the number of files
    num_files = len(files)
    dic_file_cnt[item_path[-4:]] = num_files
    
dic_file_cnt = dict(sorted(dic_file_cnt.items()))
pd.DataFrame(dic_file_cnt, index=["每天有資訊的站點"]).T
```

```python
with open("../data/測站.json", 'r') as f:
    test_station = json.load(f)
```

```python
stations = list(data.values())
t_s = list(test_station.values())
for s in stations:
    l = [(s["lat"] - t["lat"]) ** 2 + (s["lng"] - t["lng"]) ** 2 for t in t_s]
    s["ts"] = (t_s[l.index(min(l))]["name"])
    s["ts_id"] = (t_s[l.index(min(l))]["id"])
```

```python
dic = defaultdict(int)
for s in stations:
    dic[s["t_s"]] += 1
dic
```

```python
new_demo = {}
keys = list(data.keys())
for i in range(len(stations)):
    new_demo[keys[i]] = stations[i]
```

```python
with open("../data/new_demographic.json", 'w', encoding="utf-8") as json_file:
    json.dump(new_demo, json_file, indent=2, ensure_ascii=False)
```
